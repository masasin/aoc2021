```python
from aocd.models import Puzzle
import dotenv
import numpy as np
import matplotlib.pyplot as plt
dotenv.load_dotenv();
```

```python
puzzle = Puzzle(2021, 6)
data = [int(i) for i in puzzle.input_data.split(",")]
```

```python
fish = data[:]
counts = [len(fish)]
for i in range(80):
    to_add = 0
    for i in range(len(fish)):
        if fish[i] == 0:
            fish[i] = 6
            to_add += 1
        else:
            fish[i] -= 1
    fish.extend([8] * to_add)
    counts.append(len(fish))
puzzle.answer_a = len(fish)
```

```python
fig, [ax_normal, ax_semilog] = plt.subplots(ncols=2, figsize=(12, 6))
ax_normal.plot(counts)
ax_semilog.semilogy(counts)
```

```python
fish = np.array([data.count(i) for i in range(9)])
counts = [fish.sum()]
for _ in range(256):
    fish = np.roll(fish, -1)
    fish[6] += fish[8]
    counts.append(fish.sum())
puzzle.answer_b = fish.sum()
```

```python
fig, [ax_normal, ax_semilog] = plt.subplots(ncols=2, figsize=(12, 6))
ax_normal.plot(counts)
ax_semilog.semilogy(counts)
```
